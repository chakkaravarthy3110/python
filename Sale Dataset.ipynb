```python
import numpy as np

categories = np.array([
    "Beverages",
    "Snacks",
    "Dairy",
    "Bakery",
    "Frozen",
    "Household",
    "Personal Care"
])

quantity = np.array([120, 150, 100, 80, 90, 110, 95])

unit_price = np.array([1.5, 2.0, 3.5, 2.5, 4.0, 5.0, 6.0])

profit_per_item = np.array([0.5, 0.7, 1.0, 0.8, 1.2, 1.5, 2.0])

total_sales = quantity * unit_price

print("===== TOTAL SALES PER CATEGORY =====")
for i in range(len(categories)):
    print(categories[i], ":", total_sales[i])

total_revenue = np.sum(total_sales)
average_sales = np.mean(total_sales)

print("\nTotal Revenue:", total_revenue)
print("Average Sales:", average_sales)

mean_sales = np.mean(total_sales)
median_sales = np.median(total_sales)
variance_sales = np.var(total_sales)
std_sales = np.std(total_sales)

print("\n===== STATISTICAL ANALYSIS =====")
print("Mean:", mean_sales)
print("Median:", median_sales)
print("Variance:", variance_sales)
print("Standard Deviation:", std_sales)

print("\n===== INDEXING & SLICING =====")

dairy_index = np.where(categories == "Dairy")
print("Dairy Sales:", total_sales[dairy_index])

print("First 3 Categories Sales:", total_sales[:3])

high_sales = total_sales > 300
print("Categories with Sales > 300:", categories[high_sales])

total_profit = quantity * profit_per_item
overall_profit = np.sum(total_profit)

print("\n===== PROFIT ANALYSIS =====")
for i in range(len(categories)):
    print(categories[i], "Profit:", total_profit[i])

print("Overall Profit:", overall_profit)

max_index = np.argmax(total_sales)
min_index = np.argmin(total_sales)

print("\n===== PERFORMANCE ANALYSIS =====")
print("Highest Revenue Category:", categories[max_index])
print("Lowest Revenue Category:", categories[min_index])


print("\n===== INTERPRETATION =====")
print("• The average sales value indicates overall sales performance.")
print("• Standard deviation shows how much sales vary between categories.")
print("• Highest revenue category contributes most to total income.")
print("• Profit analysis helps identify most profitable product line.")
```
